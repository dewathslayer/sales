```python
%%sql
--to find the total net revenue by categories in 2022 and 2023

SELECT
product.categoryname,
SUM(CASE WHEN EXTRACT(year FROM orderdate)=2022 THEN sales.netprice*sales.quantity*sales.exchangerate END) AS net_revenue_2022,
SUM(CASE WHEN EXTRACT(year FROM orderdate)=2023 THEN sales.netprice*sales.quantity*sales.exchangerate END) AS net_revenue_2023
FROM
sales
LEFT JOIN product ON product.productkey=sales.productkey
GROUP BY
product.categoryname
```
